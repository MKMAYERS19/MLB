```python
# How many times has each team won the World Series?
Number_of_wins_per_team = World_Series.groupby('Winners').count()
Number_of_wins_per_team = Number_of_wins_per_team.sort_values(by=['Season'], ascending = False)
Number_of_wins_per_team = Number_of_wins_per_team.drop(['Winners Score', 'Losers Scores', 'Losers'], axis=1)
Number_of_wins_per_team.astype(int)
```

```python
# How many times has each team lost the Worls Series?
Number_of_loses_per_team = World_Series.groupby('Losers').count()
Number_of_loses_per_team = Number_of_loses_per_team.sort_values(by=['Season'], ascending = False)
Number_of_loses_per_team = Number_of_loses_per_team.drop(['Winners','Winners Score', 'Losers Scores'], axis=1)
Number_of_loses_per_team.astype(int)
```

```python
# Plot Histogram
plt.hist(Number_of_wins_per_team['Season'])
plt.xlabel('Number of World Series Titles')
plt.ylabel('Number of Teams')
plt.title('Distribution of World Series titles')
```

```python
# Join DataFrames
overall = pd.concat([Number_of_wins_per_team, Number_of_loses_per_team], axis=1, join='outer', sort=True)

# Rename Columns
overall.columns = ['Wins', 'Loses']

# Fill Null values
overall = overall.fillna(0)

overall
```

```python
# Win percentage 
Percentage_Win = (overall['Wins'] / (overall['Loses'] + overall['Wins'])) * 100
Percentage_Win
overall['Win Percentage'] = Percentage_Win

# Total appearances/League Championships 
League_Championships = overall['Wins'] + overall['Loses']
overall['League Championships'] = League_Championships
overall

# Fill clean data
overall = overall.fillna(0)
overall.astype(int)
```

```python
Divisions = Divisions.astype(str)
Divisions = Divisions.apply(lambda x: x.str.strip())
Divisions
```

```python
# American League West Winners 
AL_West_Titles = Divisions.groupby('AL West').count()
AL_West_Titles = AL_West_Titles.sort_values(by=['Season'], ascending = False)
AL_West_Titles = AL_West_Titles.drop(['NL East', 'AL East', 'NL West','NL Central', 'AL Central'], axis=1)
AL_West_Titles
```

```python
# American League Central Winners 
AL_Central_Titles = Divisions.groupby('AL Central').count()
AL_Central_Titles = AL_Central_Titles.sort_values(by=['Season'], ascending = False)
AL_Central_Titles = AL_Central_Titles.drop(['NL East', 'AL East', 'NL West','NL Central', 'AL West'], axis=1)
AL_Central_Titles = AL_Central_Titles.drop(AL_Central_Titles.index[0])
AL_Central_Titles
```

```python
# American League East Winners
AL_East_Titles = Divisions.groupby('AL East').count()
AL_East_Titles = AL_East_Titles.sort_values(by=['Season'], ascending = False)
AL_East_Titles = AL_East_Titles.drop(['NL East', 'AL West', 'NL West','NL Central', 'AL Central'], axis=1)
AL_East_Titles
```

```python
# National League West Winners
NL_West_Titles = Divisions.groupby('NL West').count()
NL_West_Titles = NL_West_Titles.sort_values(by=['Season'], ascending = False)
NL_West_Titles = NL_West_Titles.drop(['NL East', 'AL East', 'AL West','NL Central', 'AL Central'], axis=1)
NL_West_Titles
```

```python
# National League Central Winners 
NL_Central_Titles = Divisions.groupby('NL Central').count()
NL_Central_Titles = NL_Central_Titles.sort_values(by=['Season'], ascending = False)
NL_Central_Titles = NL_Central_Titles.drop(['NL East', 'AL East', 'NL West','AL Central', 'AL West'], axis=1)
NL_Central_Titles = NL_Central_Titles.drop(NL_Central_Titles.index[0])
NL_Central_Titles
```

```python
# National League East Winners
NL_East_Titles = Divisions.groupby('NL East').count()
NL_East_Titles = NL_East_Titles.sort_values(by=['Season'], ascending = False)
NL_East_Titles = NL_East_Titles.drop(['AL East', 'AL West', 'NL West','NL Central', 'AL Central'], axis=1)
NL_East_Titles
```

```python
# Combine all these winners into one table
# Join DataFrames
Titles = pd.concat([AL_West_Titles, AL_Central_Titles], axis=1, join='outer', sort=True)
Titles = pd.concat([Titles, AL_East_Titles], axis=1, join='outer', sort=True)
Titles = pd.concat([Titles, NL_West_Titles], axis=1, join='outer', sort=True)
Titles = pd.concat([Titles, NL_Central_Titles], axis=1, join='outer', sort=True)
Titles = pd.concat([Titles, NL_East_Titles], axis=1, join='outer', sort=True)

# Rename Columns
Titles.columns = ['AL West Wins', 'AL Central Wins', 'AL East Wins', 'NL West Wins', 'NL Central Wins', 'NL East Wins']

# Fill Null values
Titles = Titles.fillna(0)
Titles = Titles.astype(int)
Titles
```

```python
Total = Titles['AL West Wins'] + Titles['AL Central Wins'] + Titles['AL East Wins'] + Titles['NL West Wins'] + Titles['NL Central Wins'] + Titles['NL East Wins']
Titles['Total'] = Total
Titles
```

```python
Summary = pd.concat([overall, Titles], axis=1, join = 'outer', sort=True)
Summary = Summary.fillna(0)
Summary = Summary.drop(['Loses','Win Percentage', 'AL West Wins', 'AL Central Wins', 'AL East Wins', 'NL West Wins', 'NL Central Wins', 'NL East Wins'], axis = 1)
Summary.columns = ['World Series Wins', 'League Championships', 'Division Titles']
Summary.drop(Summary.index[-1])
```
